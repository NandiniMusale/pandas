```python
df_sales=pd.read_csv(r'C:\Users\DELL\Downloads\sales_data_final.csv')
df_sales.head()
```

```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(23)
```

```python
df_sales['profit_flag']=df_sales['profit']
df_sales
```

```python
df_sales['profit_flag']=df_sales['profit'].apply(profit_flag)
df_sales
```

```python
df_sales['cust_name_len']=df_sales['customer_name'].apply(len)
df_sales
```

```python
def quantity_flag(a):
    if a<=10:
        return 'low'
    if a>10 and a<=20:
        return 'medium'
    if a>20:
        return 'high'
```

```python
quantity_flag(20)
```

```python
df_sales['quantity_flag']=df_sales['quantity'].apply(quantity_flag)
df_sales
```

```python
df_sales['squre-qunatity']=df_sales['quantity'].apply(lambda a :a**2)
df_sales
```
